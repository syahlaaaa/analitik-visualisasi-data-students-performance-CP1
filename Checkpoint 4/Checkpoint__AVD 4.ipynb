#***Outliers***

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Insight

Berdasarkan hasil pengecekan, jumlah outliers di setiap kolom numerik sangat sedikit dan tidak signifikan. Artinya, sebagian besar data berada dalam rentang yang normal dan tidak ada nilai yang menyimpang terlalu jauh dari distribusi utamanya.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Saya melakukan penanganan ini supaya data menjadi lebih stabil dan tidak terlalu dipengaruhi oleh nilai ekstrem. Tujuannya agar hasil analisis selanjutnya lebih akurat dan lebih mencerminkan kondisi sebenarnya.

Saya juga tidak ingin menghapus data, karena jumlah outlier tidak terlalu banyak dan datanya masih bisa digunakan.


```python
columns_to_clip = [
    "Study Time per Week",
    "Daily Sleep Duration",
    "Reading Score",
    "Writing Score"
]
for col in columns_to_clip:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

*Penjelasan*

Setelah menggunakan metode IQR dan clip, nilai yang terlalu ekstrem tidak dihapus, tetapi hanya disesuaikan ke batas yang masih normal.

Hasilnya, hampir semua kolom sudah tidak memiliki outlier lagi dan jumlah data tetap sama. Data jadi lebih rapi, lebih stabil.

Untuk kolom Math Score, sebelum dilakukan penanganan memang ada sedikit persentase outlier. Setelah dilakukan metode clip, ternyata persentasenya masih tetap ada dan tidak berubah.

mungkin karena nilai Math Score memang dari awal sudah berada dalam rentang yang cukup stabil. Bisa jadi nilai ekstremnya sangat sedikit atau masih berada dalam batas yang dianggap wajar berdasarkan perhitungan IQR diatas.

#Bubble

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x='Study Time per Week',
    y='Math Score',
    size='Reading Score',
    hue='Gender',
    data=df,
    sizes=(50, 1000),
    alpha=0.7,
    palette='viridis'
)

plt.title('Jam Belajar vs Nilai Matematika (Ukuran: Nilai Reading, Warna: Gender)', fontsize=16)
plt.xlabel('Jam Belajar per Minggu', fontsize=12)
plt.ylabel('Nilai Matematika', fontsize=12)
plt.grid(True)
plt.legend(title='Gender', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Namun hubungannya tidak terlalu kuat karena masih ada beberapa siswa yang jam belajarnya tidak terlalu tinggi tetapi nilainya tetap bagus. Ukuran bubble yang menunjukkan nilai reading juga terlihat beragam, dan beberapa siswa dengan nilai reading yang tinggi juga memiliki nilai matematika yang cukup baik.

Tindakan:

Dari hasil ini, siswa bisa didorong untuk lebih konsisten dalam mengatur waktu belajar setiap minggu. Selain itu, sekolah juga bisa memberikan bimbingan belajar tambahan untuk siswa yang jam belajarnya masih rendah agar nilai mereka bisa lebih meningkat.

#Heatmap

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df[['Study Time per Week','Daily Sleep Duration','Math Score','Reading Score','Writing Score']].corr(),
    annot=True,
    cmap='viridis',
    fmt='.2f'
)

plt.title('Korelasi Antar Variabel')
plt.show()
```

Insight:

Dari heatmap ini saya lihat bahwa hubungan antar variabel seperti jam belajar, durasi tidur, dan nilai siswa sebenarnya tidak terlalu kuat. Nilai korelasinya mendekati 0, yang berarti hubungan antar variabel tersebut tidak terlalu berpengaruh langsung satu sama lain.

Tindakan:

Dari sini bisa dipahami bahwa nilai siswa tidak hanya dipengaruhi oleh satu faktor saja. Jadi sekolah atau guru bisa melihat faktor lain seperti metode belajar, lingkungan belajar, atau motivasi siswa.
