#***Outliers***

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Insight

Berdasarkan hasil pengecekan, jumlah outliers di setiap kolom numerik sangat sedikit dan tidak signifikan. Artinya, sebagian besar data berada dalam rentang yang normal dan tidak ada nilai yang menyimpang terlalu jauh dari distribusi utamanya.

#***4 Pilar EDA***

#Comparison

```python
rata_gender = df.groupby('Gender')[['Math Score','Reading Score','Writing Score']].mean()

rata_gender.plot(kind='bar', figsize=(8,5))
plt.title('Rata-rata Nilai Berdasarkan Gender')
plt.xlabel('Gender')
plt.ylabel('Rata-rata Nilai')
plt.xticks(rotation=0)
plt.show()

```

Insight

Berdasarkan grafik rata-rata nilai berdasarkan gender, terlihat bahwa perbedaannya tidak terlalu jauh antara siswa perempuan dan laki-laki. Nilai matematika keduanya hampir sama, berada di kisaran 69–70. Untuk reading dan writing, siswa laki-laki sedikit lebih tinggi dibandingkan perempuan, tetapi selisihnya sangat kecil.

Secara keseluruhan, bisa disimpulkan bahwa gender tidak menunjukkan perbedaan yang signifikan terhadap performa akademik dalam dataset ini karena rata-rata nilainya hampir seimbang.

#*Composition*

```python
total_nilai = df[['Math Score','Reading Score','Writing Score']].sum().sort_values(ascending=False)
top3_nilai = total_nilai.head(3)

plt.figure(figsize=(8,6))
top3_nilai.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Proporsi Total Nilai per Mata Pelajaran')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

Insight

Berdasarkan pie chart, terlihat bahwa proporsi total nilai dari ketiga mata pelajaran hampir seimbang. Math Score memiliki proporsi paling besar yaitu sekitar 34.3%, diikuti oleh Writing Score sebesar 33.0%, dan Reading Score sebesar 32.7%.

Selisihnya sangat kecil, sehingga bisa disimpulkan bahwa performa siswa di ketiga mata pelajaran relatif merata dan tidak ada satu mata pelajaran yang mendominasi secara signifikan.

#*Distribution*

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Math Score'], bins=20, kde=True)
plt.title('Distribusi Nilai Matematika')
plt.xlabel('Math Score')
plt.ylabel('Frekuensi')
plt.show()
```

Insight

Nilai matematika siswa cenderung terkonsentrasi di angka 60–80 dengan rata-rata sekitar 70. Distribusinya cukup normal dan tidak menunjukkan penyimpangan yang ekstrem.

#*Relationship*

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Math Score','Reading Score','Writing Score']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('Korelasi antar Mata Pelajaran')
plt.show()
```

Insight

Berdasarkan heatmap korelasi, nilai antar mata pelajaran hampir tidak memiliki hubungan. Hal ini terlihat dari angka korelasi yang sangat kecil (mendekati 0). Artinya, nilai matematika, reading, dan writing tidak saling memengaruhi secara signifikan. Siswa yang tinggi di satu mata pelajaran belum tentu tinggi di mata pelajaran lainnya.
